### Apercu generator

```python
def sample_generator_input(batch_size):
    # Generator inputs
    sampled_noise = np.random.normal(0, 1, (batch_size, 62))
    sampled_labels = np.random.randint(0, 10, batch_size).reshape(-1, 1)
    sampled_labels = to_categorical(sampled_labels, num_classes=10)

    return sampled_noise, sampled_labels
```

```python
model = Sequential()

model.add(Dense(128 * 7 * 7, activation="relu", input_dim=72))
model.add(Reshape((7, 7, 128)))
model.add(BatchNormalization(momentum=0.8))
model.add(UpSampling2D())
model.add(Conv2D(128, kernel_size=3, padding="same"))
model.add(Activation("relu"))
model.add(BatchNormalization(momentum=0.8))
model.add(UpSampling2D())
model.add(Conv2D(64, kernel_size=3, padding="same"))
model.add(Activation("relu"))
model.add(BatchNormalization(momentum=0.8))
model.add(Conv2D(1,kernel_size=3, padding='same'))
model.add(Activation("tanh"))

gen_input = Input(shape=(72,))
img = model(gen_input)

gennerator = Model(gen_input, img)
```

```python
gennerator.load_weights("models/gans/infogan/generator_weights.hdf5")
```

```python
fig, axs = plt.subplots(10, 20)
fig.set_size_inches(18.5, 10.5)

for l in range(10):

    sampled_noise, _ = sample_generator_input(20)
    label = to_categorical(np.full(fill_value=l, shape=(20,1)), num_classes=10)
    gen_input = np.concatenate((sampled_noise, label), axis=1)

    gen_imgs = gennerator.predict(gen_input)
    gen_imgs = 0.5 * gen_imgs + 0.5


    for x in range(20):
        axs[l,x].imshow(gen_imgs[x,:,:,0], cmap='gray')
        axs[l,x].axis('off')
```

## Recuperation du discriminateur

```python
# Load the dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Rescale -1 to 1
train_images = (train_images.astype(np.float32) - 127.5) / 127.5
test_images = (test_images.astype(np.float32) - 127.5) / 127.5

train_images = np.expand_dims(train_images, axis=3)
test_images = np.expand_dims(test_images, axis=3)

test_labels_normal = to_categorical(test_labels)

x, y = shuffle(train_images, train_labels, random_state=15)
x_train = x[0:100]
y_train = y[0:100].reshape(-1, 1)
y_train = to_categorical(y_train)
```

```python
dic = {0:7, 1:5, 2:8, 3:1, 4:3, 5:9, 6:4, 7:0, 8:2, 9:6}
y_new = [dic.get(n, n) for n in y[0:100]]

test_labels_inverse = to_categorical([dic.get(n, n) for n in test_labels])
test_labels_normal = to_categorical(test_labels)
y_train_inverse = to_categorical(y_new) 
```

### Sans augmentation

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(28,28,1)))
model.add(Activation('relu'))
model.add(BatchNormalization(axis=-1))
model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization(axis=-1))
model.add(Conv2D(64,(3, 3)))
model.add(Activation('relu'))
model.add(BatchNormalization(axis=-1))
model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
q_net = Dense(256, activation='relu')(model.layers[-1].output)
dp = Dropout(0.5)(q_net)
label = Dense(10, activation='softmax')(dp)
discriminator = Model(model.input, label)
discriminator.load_weights("models/gans/infogan/auxilliary.hdf5")
```

```python
discriminator.summary()
```

```python
for l in discriminator.layers[:-2]:
    print(l)
    l.trainable = False
```

```python
mcp = ModelCheckpoint("models/infogan_best.h5", monitor='val_acc', verbose=0, save_best_only=True, mode='auto')
```

```python
discriminator.compile(loss='categorical_crossentropy', optimizer = "adam", metrics = ["accuracy"])

history = discriminator.fit(x_train, y_train_inverse, steps_per_epoch=32, validation_steps=32, validation_data=(test_images, test_labels_inverse), epochs=200, callbacks=[mcp])

```

### Avec augmentation

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(28,28,1)))
model.add(Activation('relu'))
model.add(BatchNormalization(axis=-1))
model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization(axis=-1))
model.add(Conv2D(64,(3, 3)))
model.add(Activation('relu'))
model.add(BatchNormalization(axis=-1))
model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
q_net = Dense(256, activation='relu')(model.layers[-1].output)
dp = Dropout(0.5)(q_net)
label = Dense(10, activation='softmax')(dp)
discriminator = Model(model.input, label)
discriminator.load_weights("models/gans/infogan/auxilliary.hdf5")
```

```python
for l in discriminator.layers[:-3]:
    print(l)
    l.trainable = True
```

```python
mcp = ModelCheckpoint("models/infogan_aug_best.h5", monitor='val_acc', verbose=0, save_best_only=True, mode='auto')
discriminator.compile(loss='categorical_crossentropy', optimizer = "adam", metrics = ["accuracy"])
datagen = ImageDataGenerator(
        rotation_range=13,
        zoom_range = 0.1,
        width_shift_range=0.13,  
        height_shift_range=0.13)

datagen.fit(x_train)
```

```python
history2 = discriminator.fit_generator(datagen.flow(x_train,y_train_inverse, batch_size=32),
                    epochs = 100, callbacks=[mcp],
                    validation_data = (test_images,test_labels_inverse),
                    verbose = 1, steps_per_epoch=15)
```

```python
test_model = load_model("models/infogan_aug_best.h5")
test_loss, test_acc = test_model.evaluate(test_images, test_labels_inverse)
test_acc
```
